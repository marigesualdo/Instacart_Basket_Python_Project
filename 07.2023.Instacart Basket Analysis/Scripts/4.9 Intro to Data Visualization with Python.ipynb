# 3. Creating line charts

```python
# As it will need a lot of memory power to run this line chart, it is necessary to do it with a sample, of for example 2.000.000 
# rows. To do this with reliable results it is necessary to make sampling!
```

```python
# Spliting the dataframe in 70/30 percent.
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merged)) <= 0.7
```

```python
# Create a list holding true/false to test np.random.rant () <= 0.7
dev = np.random.rand(len(ords_prods_merged)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big
big = ords_prods_merged [dev]
```

```python
# Store 30% of the sample in the dataframe small
small = ords_prods_merged [~dev]
```

```python
# Checking the results
len(ords_prods_merged)
```

```python
len(big)+len(small)
```

```python
# reducing your samples to only those columns necessary for the chart, namely “prices” and “orders_day_of_week.”
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Creating the line chart
linechart = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Exporting as a figure
linechart.figure.savefig (os.path.join (path_2, 'linechart_orders_dow.png'))
```
