# 3. Creating variables for customer profile including "Single adult" or "Young Parent"

```python
# Determining Age, Income and Dependants variables
# Age:

high_act_cust ['Age'].describe()
```

```python
## The mean age from the customers groups is 49.5; the age range varies between 18 and 81. 
## The minimum age is 18, the maximum age 81
## As the quartil interval is divided in 33 (25%) and 65 (75%), the age groups will be created as follows:
# Young age = 0 - 33
# Middle age = 33 - 65
# Old age = over 65
```

```python
# Determining flags based on Age

high_act_cust.loc[high_act_cust['Age'] < 33, 'age_flag'] = 'Young age' 
```

```python
high_act_cust.loc[(high_act_cust['Age'] >=33 ) & (high_act_cust['Age'] <65), 'age_flag'] = 'Middle age' 
```

```python
high_act_cust.loc[high_act_cust['Age'] >= 65, 'age_flag'] = 'Old age' 
```

```python
high_act_cust ['age_flag'].value_counts (dropna = False)
```

```python
# Income

high_act_cust ['income'].describe()
```

```python
## The mean income from the customers groups is US$ 99.802 ; the income range varies between US$ 25903 and US$ 59390. 
## The minimum income is US$ 25903, the maximum income US$ 59390
## As the quartil interval is divided in US$ 67440 (25%) and US$ 128160 (75%), the income groups will be created as follows:
# Low income = 0 - 67440
# Middle income = 67440 - 128160
# High income = over 128160
```

```python
# Determining flags based on income
high_act_cust.loc[high_act_cust['income'] < 67440, 'income_flag'] = 'Low income' 
```

```python
high_act_cust.loc[(high_act_cust['income'] >= 67440 ) & (high_act_cust['income'] <128160), 'income_flag'] = 'Middle income' 
```

```python
high_act_cust.loc[high_act_cust['income'] >= 128160, 'income_flag'] = 'High income' 
```

```python
high_act_cust ['income_flag'].value_counts (dropna = False)
```

```python
high_act_cust.head()
```

```python
# Dependants:

high_act_cust ['n_dependants'].describe()
```

```python
# The mean of dependants is =~ 2; the range varies between 0 and 3 children. To facilitate the column construction,
# I will estabilish 2 categories, no parent or parent.
```

```python
high_act_cust.loc[high_act_cust['n_dependants'] >= 1, 'parent_flag'] = 'Parent' 
```
