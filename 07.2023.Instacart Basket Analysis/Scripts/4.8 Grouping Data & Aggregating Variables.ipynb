## 1.2 Aggregating Data with transform()

```python
# Used to create a flag - for example: a loyalty flag
# Flags help target specific groups within your data according to a set of conditions, allowing for segmented analysis.
```

```python
#locating these loyalty customers at Instacart using the transform () function
# Criteria:
# If the maximum orders > 40, then ‚ÄúLoyal customer.‚Äù
# If the maximum orders > 10 or <= 40, then ‚ÄúRegular customer.‚Äù
# If the maximum <= 10, then  ‚ÄúNew customer.‚Äù
```

```python
#1. Split the data into groups based on the ‚Äúuser_id‚Äù column.
#2. Apply the transform() function on the ‚Äúorder_number‚Äù column to generate the maximum orders for each user.
#3. Create a new column, ‚Äúmax_order,‚Äù into which you‚Äôll place the results of your aggregation.

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking results: 
ords_prods_merge.head(15)
```

```python
# The column max order () means the maximum number of orders made by each user ID
```

```python
ords_prods_merge.head(100)
```

## 1.3 Deriving Columns with loc()

```python
#Creating a flag that assigns a ‚Äúloyalty‚Äù label to a user ID based on its corresponding max order value
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Frequency of loyalty flag column
ords_prods_merge ['loyalty_flag'].value_counts (dropna = False)
```

```python
# limiting the output to see only the columns that we need.
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
# Question 6.Create a spending flag for each user based on the average price across all their orders using the following criteria:

#Creating a flag that assigns a ‚Äúloyalty‚Äù label to a user ID based on its corresponding price value

# First it is necessary to create a new column with the price means by customer.
```

```python
# Question 7. Create an order frequency flag that marks the regularity of a user‚Äôs ordering behavior
```

```python
# First it is necessary to create a new column with the median in the ‚Äúdays_since_prior_order‚Äù by customer.
ords_prods_merge['user_median_days_since_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head (15)
```

```python
# Making the flags with the loc function.
ords_prods_merge.loc[ords_prods_merge['user_median_days_since_order'] > 20, 'mediandays_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['user_median_days_since_order'] <= 20) & (ords_prods_merge ['user_median_days_since_order'] > 10), 'mediandays_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_median_days_since_order'] <= 10, 'mediandays_flag'] = 'Frequent customer'
```

```python
# Frequency of the price flag column
ords_prods_merge ['mediandays_flag'].value_counts (dropna = False)
```
